```python
from markitdown import MarkItDown

# 도구 객체 생성
md = MarkItDown()

# 파일 변환 (노트북이 위치한 폴더에 sample.pdf가 있다고 가정)
result = md.convert("업무속도를 2배로 높이는 AI 업무 활용법 (이임복) (z-library.sk, 1lib.sk, z-lib.sk).pdf")

# 결과가 마크다운 텍스트로 잘 변환되었는지 확인
print(result.markdown)


# 예시: 변환된 내용을 다시 파일로 저장
with open("converted_file.md", "w", encoding="utf-8") as f:
    f.write(result.markdown)

# 예시: 내용의 일부분만 추출
print(result.markdown[:500]) # 앞의 500자만 미리보기

```

```python
from markitdown import MarkItDown
from openai import OpenAI

# OpenAI 모델을 연결하여 이미지/스캔본 해석
md = MarkItDown(
    llm_client=OpenAI(),
    llm_model="gpt-4o"
)
result = md.convert("handwritten_note.jpg")
print(result.markdown)
```
